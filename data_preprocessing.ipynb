```python
import pandas as pd
```

```python
strava = pd.read_excel("data/strava_running_data.xlsx")
strava.head()
```

```python
strava["distance_km"] = strava["distance"] / 1000
strava["duration_min"] = strava["moving_time"] / 60
strava["pace_min_per_km"] = strava["duration_min"] / strava["distance_km"]
```

```python
strava["start_dt"] = pd.to_datetime(strava["start_date_local"])
strava["hour"] = strava["start_dt"].dt.hour
strava["weekday"] = strava["start_dt"].dt.weekday
strava.head()
```

```python
strava.to_csv("data/strava_run_data.csv", index=False)
```

```python
import numpy as np
```

```python
strava = pd.read_csv("data/strava_run_data.csv")
strava.head()
```

```python
strava = strava.replace([np.inf, -np.inf], np.nan)
strava = strava[strava["pace_min_per_km"].notnull()]
strava.shape
```

```python
q_fast = strava["pace_min_per_km"].quantile(0.33)
q_med  = strava["pace_min_per_km"].quantile(0.66)
print("Pace quantiles:")
print("  fast/medium boundary :", q_fast)
print("  medium/slow boundary :", q_med)
```

```python
def pace_group_assignment(pace):
    if pace <= q_fast:
        return "fast"
    elif pace <= q_med:
        return "medium"
    else:
        return "slow"

strava["pace_group"] = strava["pace_min_per_km"].apply(pace_group_assignment)
strava.head()
```

```python
playlist_groups = {"fast": "5ymkuBk3C1Iu4KTdJBK9vy", "medium": "7pWNfjCrJmS4ILho1S4CvK", "slow": "5Egmw4o1GAN8QkhXGaxao6"}
strava["playlist_id"] = strava["pace_group"].map(playlist_groups)
```

```python
activity_mapping = strava[["Sr. no.", "pace_group", "playlist_id"]].copy()
activity_mapping.head()
```

```python
activity_mapping[["Sr. no.", "playlist_id"]].to_csv("data/activity_playlist_correlation.csv", index=False)
```

```python
strava = pd.read_csv("data/strava_run_data.csv")
activity = pd.read_csv("data/activity_playlist_correlation.csv")
playlists = pd.read_csv("data/playlist_metadata.csv")
```
